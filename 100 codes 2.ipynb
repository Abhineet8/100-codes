(5.)  User will input (3ages).Find the oldest one.

```python
age1 = int(input("Input first age "))
age2 = int(input("Input second age "))
age3 = int(input("Input third age "))

agemax=max(age1,age3,age2)
print('The oldest age',agemax)
```

(7.) User will input (2numbers).Write a program to swap the numbers

```python
num1 = int(input("Input first num "))
num2 = int(input("Input second num "))
swapper = num1
num1 = num2
num2 = swapper
print("value of num1: ", num1)
print("value of num2: ", num2)
```
